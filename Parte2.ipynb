Decisao importante: aqui sera feita a categorizacao dos aspectos financeiros e de faixa etaria das ZCTAs.

Catgorizacao por idades:
    Criancas -> 0 ate 9 anos
    Jovens -> 10 ate 19 anos
    Adultos -> 20 ate 59 anos 
    Idosos -> 60+ anos

Categorizacao por capacidade financeira
    Classe Baixa: 0 a 34.999
    Classe Media: 35.000 a 99.999
    Classe Alta: 100.000 +



Desta forma, as 22 colunas que contem estes dados serao reduzidas, formando assim 7 colunas


```python
df_merged2["qtd criancas"] = df_merged2["Population_Under5Years"] + df_merged2["Population_5to9Years"] 
df_merged2["qtd Jovens"] = df_merged2["Population_10to14Years"] + df_merged2["Population_15to19Years"] 
df_merged2["qtd Adultos"] = df_merged2["Population_20to24Years"] + df_merged2["Population_25to34Years"]  + df_merged2["Population_35to44Years"]  + df_merged2["Population_45to54Years"]  + df_merged2["Population_55to59Years"] 
df_merged2["qtd Idosos"] = df_merged2["Population_60to64Years"] + df_merged2["Population_65to74Years"] + df_merged2["Population_75to84Years"]  + df_merged2["Population_85YearsAndOver"]

df_merged2["qtd classe baixa"] = df_merged2["TotalHouseholds_LessThan$10.000"] + df_merged2["TotalHouseholds_$10.000to$14.999"] + df_merged2["Population_35to44Years"]  + df_merged2["TotalHouseholds_$15.000to$24.999"]  + df_merged2["TotalHouseholds_$25.000to$34.999"] 
df_merged2["qtd classe media"] = df_merged2["TotalHouseholds_$35.000to$49.999"] + df_merged2["TotalHouseholds_$50.000to$74.999"] + df_merged2["TotalHouseholds_$75.000to$99.999"]  + df_merged2["TotalHouseholds_$15.000to$24.999"]  + df_merged2["TotalHouseholds_$25.000to$34.999"] 
df_merged2["qtd classe alta"] = df_merged2["TotalHouseholds_$100.000to$149.999"] + df_merged2["TotalHouseholds_$150.000to$199.999"] +  df_merged2["TotalHouseholds_$200.000OrMore"]
```
